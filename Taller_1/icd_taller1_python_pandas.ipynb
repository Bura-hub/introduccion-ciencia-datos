# **MÓDULO 0: PRIMEROS PASOS CON PYTHON**

Es común usar Excel para analizar datos, pero si el número de datos es grande, Excel se queda corto. En este caso, es conveniente movernos hacia una herramienta más sofisticada, como Python (el lenguaje de programación que usaremos en el curso). Entonces, antes de analizar datos complejos, debemos entender Python. Por fortuna, es un lenguaje muy legible, muy parecido a Matlab.


**0.1 Variables:** En Excel, guardamos datos en celdas (A1, B2). En Python, usamos variables.

```python
# --- VARIABLES ---

# Asignamos variables con el signo '='
# No necesitamos declarar el tipo (int, float); Python lo infiere.

voltaje = 220       # Entero (int)
resistencia = 10.5  # Decimal (float)
sistema_activo = True # Booleano (lógica binaria: 1 o 0)
nombre_pozo = "Pozo Norte-X" # Texto (string)

# La función print() muestra el resultado en pantalla
print("El voltaje es:", voltaje)

# La función type() nos dice qué tipo de dato es
print("Tipo de dato de resistencia:", type(resistencia))
```

**0.2 Operaciones Matemáticas y Librerías:** Al igual que Matlab, Python funciona como una calculadora científica:

```python
# --- CÁLCULOS EN PYTHON ---

# Ley de Ohm: I = V / R
corriente = voltaje / resistencia

# Potencia: P = V * I
potencia = voltaje * corriente

print("La corriente calculada es", corriente, "Amperios")
print("La potencia es:", potencia, "Watts")
```

A diferencia de Matlab, la mayoría de funciones de Python están en librerías, y antes de usarlas, debemos importar esas librerías:

```python
# --- LIBRERÍAS ---

# Calculemos el voltaje a partir de la potencia y la resistencia.
# Para ello debemos usar la función "raíz cuadrada", sqrt().
# Esta función está disponible en la librería "math", así que primero debemos importar esta librería.
# Además, podemos darle nombre un "alias" personalizado a la librería, por ejemplo "matem"

import math as matem

# Voltaje calculado: voltaje_calculado = sqrt(potencia*resistencia)
voltaje_calculado = matem.sqrt(potencia*resistencia)

print(f"El voltaje calculado es: {voltaje_calculado:.2f} Voltios") #.2f se usa para mostrar 2 decimales
```

**0.3 Estructuras de Datos: Listas y Diccionarios:**
En ingeniería rara vez trabajamos con un solo dato; trabajamos con vectores o series. En Python, usamos listas (que definimos usando corchetes []) y diccionarios (definidas usando llaves {}).

Las listas son los "vectores" de Pyhton; es decir, colecciones ordenadas indexadas por posición (0,1,2...). Por otro lado, los diccionarios son colecciones desordenadas de pares "llave"-"valor".

**Nota:** En Python, los elementos de las listas se empiezan a contar desde CERO, no desde uno.

```python
# --- LISTAS (VECTORES) ---

# Una lista de temperaturas tomadas por un sensor en 5 momentos distintos
temperaturas = [105.4, 106.2, 108.1, 107.5, 110.0]

# Acceder al primer dato (índice 0)
print("Temperatura inicial:", temperaturas[0])

# Acceder al último dato (índice -1)
print("Temperatura final:", temperaturas[-1])

# ¿Cuántos datos tenemos?
print("Número de mediciones:", len(temperaturas)) #la función len() devuelve el número de elementos de la lista

# Operaciones básicas: Agregar un dato nuevo
temperaturas.append(24.1) #el método .append() agrega un elemento al final de la lista
print("Lecturas actualizadas:", temperaturas)
```

**0.4 Identación, Condicionales (if, else) y Ciclos (for):**
A diferencia de otros lenguajes que usan llaves {} para agrupar código, Python utiliza la indentación (sangría). Es obligatorio, no estético.

Cada vez que una línea termina en dos puntos (:), la siguiente línea debe tener identación (usualmente 4 espacios). Todo el código que comparte el mismo nivel de indentación pertenece al mismo bloque lógico.

```python
# --- CONDICIONALES ---

import random # importamos la librería random

retorno_diario = -random.randint(1, 10) # simulamos una caída aleatoria entre el 10% y 1%
print(f"retorno: {retorno_diario} %")

# Noten los dos puntos (:) al final de la línea 'if'
# Noten el espacio (4 espacios o 1 TAB) en las líneas print
if retorno_diario < -5:
    print("ALERTA: Caída significativa detectada.") # esta línea se ejecuta si es TRUE porque está identada respecto al "if"
    print("Revisar noticias macroeconómicas.") # esta línea también se ejecuta si es TRUE porque está identada
else:
    print("Mercado estable.") # esta línea se ejecuta si es FALSE porque está identada respecto al "else"

print("Fin del análisis.") # Esta línea se ejecuta SIEMPRE, porque no tiene indentación.
```

```python
# --- CICLOS ---

# Caso1: Usemos un ciclo for para operar sobre una lista
precios_USD = [100, 102, 98, 105, 110] # lista de precios en dolares
precios_COP = []

# "Para cada 'precio' dentro de la lista 'precios'..."
for precio in precios_USD:
    # Lo convertimos a pesos (multiplicamos por 3650)
    precio_COP = precio*3650
    precios_COP.append(precio_COP)

print("Lista de precios en pesos es", precios_COP)
```

```python
# Caso 2: Si no tienes una lista, pero quieres repetir algo 5 veces, usas la función range()
for numero in range(5): # range(5) crea una secuencia: 0, 1, 2, 3, 4
    print(f"Vuelta número: {numero}")
```

```python
# Caso 3: Usemos ahora un ciclo for para operar sobre un diccionario
ventas = {
    "Lunes": 100,
    "Martes": 250,
    "Miércoles": 180
}

# Si queremos recorrer el diccionario sobre las LLAVES:
for dia in ventas.keys():
  print(f"El día {dia} se registraron ventas")

# Si queremos recorrer el diccionario sobre los VALORES:
total_ventas = 0
for valor in ventas.values():
  total_ventas = total_ventas + valor
print(f"El total de ventas de la semana fue {total_ventas}")
```

**Nota:** Aunque el ciclo for es fundamental para entender la lógica, en Ciencia de Datos intentamos evitarlo cuando procesamos millones de datos. Preferimos la "vectorización" (operar sobre toda la columna a la vez). Esto lo veremos en el Módulo 1 (Pandas).

# **MÓDULO 1: INTRODUCCIÓN A PANDAS**

Las listas y diccionarios de Python son útiles, pero para trabajar con datos a un nivel más avanzado necesitamos algo más eficiente. Aquí entra Pandas, una librería de Python diseñada para trabajar con datos.

Pandas introduce el DataFrame: Es una tabla con etiquetas (filas y columnas), donde los tipos de datos están optimizados para manejar millones de registros. Es como una hoja de Excel, pero eficiente y programable. Al igual que en Excel, en un DataFrame se pueden almacenar datos de diferentes tipos: números, fechas, strings, etc.


**1.1 Preliminares:**
Primero, importamos la librería y descargamos los datos con los que trabajaremos: las notas de un curso avanzado en topología variacional.

```python
# --- IMPORTAR PANDAS ---

import pandas as pd # Importamos la librería con el alias 'pd' (estándar mundial)


# --- CARGAR DATOS ---

# 1. DEFINICIÓN DE LA FUENTE DE DATOS
# ID del archivo donde están las notas:
sheet_id = "1mZJTauwhpKWSJ5d6LxH1wnJ1GNC_Dact0_5yG6p8OZg"
# Construimos la URL de exportación del archivo (el archivo está disponible en Google Drive):
url = f"https://docs.google.com/spreadsheets/d/{sheet_id}/export?format=csv"

# 2. CARGA DEL DataFrame
# Construimos nuestro DataFrame, llamado "df_notas", leyendo el archivo desde la URL:
df_notas = pd.read_csv(url)


# Veamos cómo es el DataFrame que acabamos de descargar:
display(df_notas)
```

**1.2 Exploración de Datos:**
Antes de analizar, debemos entender qué forma tienen nuestros datos:
* `.shape`: Nos dice cuántas filas (estudiantes) y columnas tenemos.
* `.describe()`: Nos da estadísticas rápidas (nota máxima, mínima, promedio por parcial).
* `.dtypes`: Nos dice qué tipo de datos tenemos (números o texto).

```python
# Ver dimensiones (Filas, Columnas)
print(f"Dimensiones de los datos: {df_notas.shape}")

# Ver tipos de datos
print("\n--- Tipos de datos ---")
print(df_notas.dtypes)

# Estadísticas básicas
print("\n--- Estadísticas del Curso ---")
# describe() nos dará la media, desviación estándar, min, max y cuartiles
display(df_notas.describe())
```

**1.3 Etiquetas:**
Las columnas y filas de los **DataFrames** tienen etiquetas que nos ayudan a identificarlas:

* **Etiquetas de columna:** Generalmente son los nombres de las variables almacenadas (ej. `Edad`, `Precio`).
* **Etiquetas de fila (Índice):** Actúan como el identificador único de cada fila. Comúnmente son números consecutivos, pero también pueden ser fechas (en series temporales) o textos específicos.

Una característica clave es que las columnas pueden almacenar **diferentes tipos de datos**. Por ejemplo, la primera columna puede contener números enteros y la segunda cadenas de texto (*strings*).

```python
# --- ETIQUETAS DEL DataFrame ---

print("\n--- NOMBRES DE COLUMNAS ---")
print(df_notas.columns) # Muestra los nombres de las variables
# En nuesto caso, los nombres de las columnas son: "nombre", "estado", "parcial 1", ...


print("\n--- ÍNDICE (ETIQUETAS DE FILAS) ---")
print(df_notas.index) # Muestra cómo están etiquetadas las filas
# En nuesto caso, la etiqueta de la filas es la posición que ocupa en la tabla (empezando por 0)
```

**1.4 Creación de Nuevas Columnas:**
Vamos a calcular la **Nota Definitiva** de cada estudiante.
Asumiremos que el promedio es la suma de los tres parciales dividida entre 3.

$$\text{Promedio} = \frac{\text{Parcial 1} + \text{Parcial 2} + \text{Parcial 3}}{3}$$

```python
# Crear la columna 'Definitiva' promediando los 3 parciales
df_notas['Definitiva'] = (df_notas['Parcial 1'] + df_notas['Parcial 2'] + df_notas['Parcial 3']) / 3

# Redondeamos a 1 decimal para que se vea mejor
df_notas['Definitiva'] = df_notas['Definitiva'].round(1)

# Mostramos el resultado con la nueva columna
print("--- Notas con Definitiva Calculada ---")
display(df_notas.head(4)) # .head(n) muestra las n primeras filas del DataFrame
```

**1.5 Filtrado de Datos:**
Ahora responderemos preguntas específicas sobre el curso:
1. ¿Quiénes perdieron la materia? (Nota < 3.0)
2. ¿Quiénes son los estudiantes de excelencia? (Nota > 4.5)

```python
# 1. Estudiantes que reprobaron (Definitiva menor a 3.0)

# Esto devuelve un DataFrame sólo con las filas que cumplen la condición:
reprobados = df_notas[df_notas['Definitiva'] < 3.0]

print(f"Número de estudiantes reprobados: {len(reprobados)}")
print("\n--- Lista de Reprobados ---")
display(reprobados[['Estudiante', 'Definitiva']]) # Solo mostramos nombre y nota


# 2. Estudiantes de Excelencia (Definitiva mayor a 4.5)
excelencia = df_notas[df_notas['Definitiva'] > 4.5]

print("\n--- Cuadro de Honor (Excelencia) ---")
display(excelencia[['Estudiante', 'Definitiva']])
```

**1.7 Consultas y Selección de Datos por Posición (.iloc):**
En Pandas podemos acceder a los datos de un DataFrame especificando la posición que ocupan. Es decir, mediante sus coordenadas.


*   *Ejemplo:* "Dame el dato que está en la fila 2, columna 0".

La sintaxis es: `df_notas.iloc[numero_fila, numero_columna]`

Recordemos que en Python los conteos empiezan en **0**.

```python
# 1. Dato específico: Fila 0 (Primer estudiante), Columna 0 (Primer parcial)
dato_posicion = df_notas.iloc[0, 0]
print(f"Dato en fila 0, col 0 (Nombre del primer alumno en la lista): {dato_posicion}")

# 2. Última fila del curso (útil cuando no sabemos cuántos hay)
print("\n--- Datos del último estudiante de la lista ---")
print(df_notas.iloc[-1]) # -1 significa "el último"

# 3. Slicing
# Las notas obtenidas por los 5 primeros estudiantes (filas 0 a 4) en los 3 parciales (columnas 2 a 4)
print("\n--- Subconjunto (Parciales de los 5 primeros estudiantes en la lista) ---")
# OJO: En .iloc, el límite superior NO se incluye. Ej: 2:5 cubre las posiciones 2, 3 y 4
display(df_notas.iloc[0:5, 2:5])
```
